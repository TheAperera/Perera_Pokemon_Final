## Team Building Function
- Requires a dataframe started for the team argument
- Example includes starting a team with one pokemon and then adding the rest via the function

```python

#I designed a small function to create small dataframes, or rather to build them for me
#Unforuntately I struggled with intializing a dataframe that can have variable name change inside the function
#My solution was simpler, I'll jsut design the list of teams and start their dataframes by introducing the first pokemon
#This way I could just create teams 

def teamBuilder (pokeList, team):
  for name in pokeList:
    pokePlace = pokedex_df[pokedex_df['Name'] == name]
    team = team.append(pokePlace)
    
  return team
```

## Starter Advantage

When I was younger, my favorite starter pokemon has always been the fire types, I wondered amongst all the starter pokemon is the best to start with? In this case, what if I started the 6th generation with a different starter?

```python
#Data Prep

#I started by selecting the pokemon using the .loc method to pull them out of their Pokedex Entry (which later became a problem, because .loc goes by entries, where this dataset includes alternative forms)
kanto_starter = kanto_dex.loc[:12]

#I realized by the second generation that I could use the .head(x) function to pull the first several entries of the pokedex by region/generation since they're always in the beginning of the games pokedex
johto_starter = johto_dex.head(9) #no mega evolutions
hoenn_starter = hoenn_dex.head(12)
sinnoh_starter = sinnoh_dex.head(9) #no mega evolutions
unova_starter = unova_dex.loc[554:562] #no mega evolutions, but I couldn't do the .head() trick because interestingly enough, Victini (Legendary) is first in that Generation 5 dataframe
kalos_starter = kalos_dex.head(9) #No Mega evolutions (but introduced them)

#All the starter pokmeon and their evolutions into a single dataframe
starterEvos = kanto_starter.append(johto_starter)
starterEvos = starterEvos.append(hoenn_starter)
starterEvos = starterEvos.append(sinnoh_starter)
starterEvos = starterEvos.append(unova_starter)
starterEvos = starterEvos.append(kalos_starter)

baseStarters = ['Charmander', 'Squirtle', 'Chikorita', 'Cyndaquil', 'Totodile', 'Treecko', 'Torchic', 'Mudkip', 'Turtwig', 'Chimchar', 'Piplup', 'Snivy', 'Tepig', 'Oshawott', 'Chespin', 'Fennekin', 'Froakie']
starterTeam = starterEvos[starterEvos['Name'] == 'Bulbasaur']
starterTeam = teamBuilder(baseStarters, starterTeam)

starterTeam = starterTeam.drop(columns=['#','Legendary'])
starterTeam

tidy = kanto_dex.melt(id_vars=['Name']).rename(columns=str.title)
```

## Casual V Competitive Teams
What difference is there between a non-competitive (casual) player’s team VS a competitive player’s team?


```python
#Data Prep
#Team Construction

############ ------------- Casuals 
andrew = ['Arcanine', 'Steelix', 'Swampert', 'Gardevoir', 'Torterra']
AndrewTeam = pokedex_df[pokedex_df['Name'] == 'Salamence'] 
AndrewTeam = teamBuilder (andrew, AndrewTeam) # my function made from earlier
#DROPPING Columns so I can hone in on values I'm interested in
AndrewTeam = AndrewTeam.drop(columns=['#','Legendary', 'Generation'])


anthony = ['Mismagius', 'Abomasnow', 'Magnezone', 'Clawitzer', 'Forretress', 'Tirtouga']
AnthonyTeam = pokedex_df[pokedex_df['Name'] == 'Tirtouga'] 
AnthonyTeam = teamBuilder (anthony, AnthonyTeam)


############ ------------- 2014
#2014 VGT Champion: Se Jun Park
ChampSJP = ['Gothitelle', 'Gardevoir', 'Gyarados', 'Talonflame', 'Garchomp']
ChampSJPTeam = pokedex_df[pokedex_df['Name'] == 'Pachirisu']
ChampSJPTeam = teamBuilder(ChampSJP, ChampSJPTeam)
ChampSJPTeam = ChampSJPTeam.drop(columns=['#','Legendary', 'Generation'])




#2014 VGT Runner Up: Jeudy Azzarelli
RunnerUpJA = ['Salamence', 'Charizard', 'Tyranitar', 'Garchomp', 'RotomMow Rotom']
RunnerUpJATeam = pokedex_df[pokedex_df['Name'] == 'Lucario']
RunnerUpJATeam = teamBuilder (RunnerUpJA, RunnerUpJATeam)
RunnerUpJATeam = RunnerUpJATeam.drop(columns=['#','Legendary', 'Generation'])


########### ------------- 2015

#2015 Video Game Tournament Champion: Ashton Cox
#This is the World Champion for the video game series that was held in 2015 (up to date with the dataset)
ChampAshton = [ 'RotomWash Rotom', 'Venusaur', 'Terrakion', 'LandorusTherian Forme', 'Milotic']
ChampionAshtonTeam = pokedex_df[pokedex_df['Name'] == 'Heatran'] 
ChampionAshtonTeam = teamBuilder (ChampAshton, ChampionAshtonTeam)
ChampionAshtonTeam = ChampionAshtonTeam.drop(columns=['#','Legendary', 'Generation'])


#Runner Up: Wolfe
#Interesting the runner up ran mega gengar
RunnerUpWolfe = ['Gengar', 'Scrafty', 'Politoed', 'Gothitelle', 'Arcanine']
RunnerUpWolfeTeam = pokedex_df[pokedex_df['Name'] == 'Amoonguss']
RunnerUpWolfeTeam = teamBuilder (RunnerUpWolfe, RunnerUpWolfeTeam)
RunnerUpWolfeTeam = RunnerUpWolfeTeam.drop(columns=['#','Legendary', 'Generation'])


```
